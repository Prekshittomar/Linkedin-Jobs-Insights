```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("skill_migration_public.csv")
```

```python
# Visualization 1: Distribution of net migration per 10K people in 2015
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='net_per_10K_2015', color='skyblue', bins=30)
plt.title('Distribution of Net Migration per 10K People in 2015')
plt.xlabel('Net Migration per 10K People')
plt.ylabel('Frequency')
plt.show()
```

```python
# Visualization 2: Boxplot of net migration per 10K people in 2016
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, y='net_per_10K_2016', color='skyblue')
plt.title('Boxplot of Net Migration per 10K People in 2016')
plt.ylabel('Net Migration per 10K People')
plt.show()
```

```python
# Visualization 3: Count of jobs by skill group category
plt.figure(figsize=(10, 6))
sns.countplot(data=df, y='skill_group_category', color='skyblue')
plt.title('Count of Jobs by Skill Group Category')
plt.xlabel('Count')
plt.ylabel('Skill Group Category')
plt.show()
```

```python
# Visualization 4: Scatter plot of net migration per 10K people in 2017 vs 2018
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='net_per_10K_2017', y='net_per_10K_2018', color='skyblue')
plt.title('Net Migration per 10K People in 2017 vs 2018')
plt.xlabel('Net Migration per 10K People in 2017')
plt.ylabel('Net Migration per 10K People in 2018')
plt.show()
```

```python
# Visualization 5: Line plot of net migration per 10K people from 2015 to 2019
plt.figure(figsize=(10, 6))
sns.lineplot(data=df[['net_per_10K_2015', 'net_per_10K_2016', 'net_per_10K_2017', 'net_per_10K_2018', 'net_per_10K_2019']])
plt.title('Trend of Net Migration per 10K People (2015-2019)')
plt.xlabel('Year')
plt.ylabel('Net Migration per 10K People')
plt.legend(labels=['2015', '2016', '2017', '2018', '2019'])
plt.show()
```

```python
# Visualization 6: Barplot of top 10 countries by net migration per 10K people in 2019
top_countries_2019 = df.groupby('country_name')['net_per_10K_2019'].mean().nlargest(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_countries_2019.values, y=top_countries_2019.index, color='skyblue')
plt.title('Top 10 Countries by Net Migration per 10K People in 2019')
plt.xlabel('Net Migration per 10K People')
plt.ylabel('Country')
plt.show()
```

```python
# Visualization 7: Pie chart of job distribution by World Bank income level
plt.figure(figsize=(10, 6))
df['wb_income'].value_counts().plot.pie(autopct='%1.2f%%', colors=['skyblue', 'lightblue', 'deepskyblue', 'dodgerblue'])
plt.title('Job Distribution by World Bank Income Level')
plt.ylabel('')
plt.show()
```

```python
# Visualization 8: Heatmap of correlation between net migration per 10K people from 2015 to 2019
plt.figure(figsize=(10, 6))
sns.heatmap(df[['net_per_10K_2015', 'net_per_10K_2016', 'net_per_10K_2017', 'net_per_10K_2018', 'net_per_10K_2019']].corr(), cmap='coolwarm', annot=True)
plt.title('Correlation Heatmap of Net Migration per 10K People (2015-2019)')
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.show()
```

```python
# Visualization 9: Violin plot of net migration per 10K people by World Bank region
plt.figure(figsize=(12, 8))
sns.violinplot(data=df, x='wb_region', y='net_per_10K_2019', palette='Blues')
plt.title('Distribution of Net Migration per 10K People by World Bank Region (2019)')
plt.xlabel('World Bank Region')
plt.ylabel('Net Migration per 10K People')
plt.xticks(rotation=90)
plt.show()
```

```python
# Visualization 10: Pairplot of net migration per 10K people from 2015 to 2019
sns.pairplot(df[['net_per_10K_2015', 'net_per_10K_2016', 'net_per_10K_2017', 'net_per_10K_2018', 'net_per_10K_2019']], palette='skyblue')
plt.suptitle('Pairplot of Net Migration per 10K People (2015-2019)')
plt.show()
```

```python
# Visualization 11: Stacked bar plot of job count by skill group category and World Bank income level
plt.figure(figsize=(12, 8))
sns.countplot(data=df, y='skill_group_category', hue='wb_income', palette='Blues')
plt.title('Job Count by Skill Group Category and World Bank Income Level')
plt.xlabel('Count')
plt.ylabel('Skill Group Category')
plt.legend(title='World Bank Income Level')
plt.show()
```

```python
# Visualization 12: Line plot of net migration per 10K people by country over the years
plt.figure(figsize=(12, 40))
for country in df['country_name'].unique():
    country_data = df[df['country_name'] == country]
    plt.plot(country_data.columns[-5:], country_data.iloc[0][-5:], label=country, marker='o')
plt.title('Net Migration per 10K People by Country (2015-2019)')
plt.xlabel('Year')
plt.ylabel('Net Migration per 10K People')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Visualization 13: Bar plot of average net migration per 10K people by skill group category
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='skill_group_category', y='net_per_10K_2019', estimator=sum, ci=None, palette='Blues')
plt.title('Average Net Migration per 10K People by Skill Group Category (2019)')
plt.xlabel('Skill Group Category')
plt.ylabel('Average Net Migration per 10K People')
plt.xticks(rotation=90)
plt.show()
```

```python
# Visualization 14: KDE plot of net migration per 10K people in 2019
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df['net_per_10K_2019'], color='skyblue', fill=True)
plt.title('Kernel Density Estimation of Net Migration per 10K People in 2019')
plt.xlabel('Net Migration per 10K People')
plt.ylabel('Density')
plt.show()
```
